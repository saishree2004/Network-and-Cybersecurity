### BASIC BIT MANIPULATION UTILITIES

```python
def permute(block, table):
    return ''.join(block[i-1] for i in table)




def xor(a, b):
    return ''.join('0' if i == j else '1' for i, j in zip(a, b))




def left_shift(bits, n):
    return bits[n:] + bits[:n]




def text_to_bits(text):
    return ''.join(format(ord(c), '08b') for c in text)




def bits_to_text(bits):
    return ''.join(chr(int(bits[i:i+8], 2)) for i in range(0, len(bits), 8))
```

### KEY SCHEDULING (ROUND KEY GENERATION)

```python
def generate_round_keys(key_64):
    key_56 = permute(key_64, PC1)
    C, D = key_56[:28], key_56[28:]
    round_keys = []
    for shift in SHIFT_SCHEDULE:
        C = left_shift(C, shift)
        D = left_shift(D, shift)
        round_key = permute(C + D, PC2)
        round_keys.append(round_key)
    return round_keys
```

### FEISTEL (F) FUNCTION

```python
def sbox_substitution(bits48):
    output = ''
    for i in range(8):
        block = bits48[i*6:(i+1)*6]
        row = int(block[0] + block[5], 2)
        col = int(block[1:5], 2)
        output += format(S_BOX[i][row][col], '04b')
    return output




def feistel(right32, round_key):
    expanded = permute(right32, E)
    xored = xor(expanded, round_key)
    substituted = sbox_substitution(xored)
    return permute(substituted, P)
```

### DES ENCRYPTION (ONE BLOCK)

```python
def des_encrypt_block(plain64, round_keys):
    permuted = permute(plain64, IP)
    L, R = permuted[:32], permuted[32:]
    
    for i in range(16):
        temp = R
        R = xor(L, feistel(R, round_keys[i]))
        L = temp
    combined = R + L
    return permute(combined, FP)
```

### DES DECRYPTION (ONE BLOCK)


```python
def des_decrypt_block(cipher64, round_keys):
    permuted = permute(cipher64, IP)
    L, R = permuted[:32], permuted[32:]
    
    for i in range(15, -1, -1):
        temp = R
        R = xor(L, feistel(R, round_keys[i]))
        L = temp
    combined = R + L
    return permute(combined, FP)
```

```python
plaintext = "HELLODES" # 8 characters = 64 bits
key = "MYDESKEY" # 8 characters = 64 bits


plain_bits = text_to_bits(plaintext)
key_bits = text_to_bits(key)


round_keys = generate_round_keys(key_bits)


cipher_bits = des_encrypt_block(plain_bits, round_keys)
decrypted_bits = des_decrypt_block(cipher_bits, round_keys)


print("Plaintext :", plaintext)
print("Cipher (bin):", cipher_bits)
print("Decrypted :", bits_to_text(decrypted_bits))
```
